# PREPROCESSING 

```python
from sklearn.preprocessing import LabelEncoder  #ML models work with numbers, not text LabelEncoder converts text labels into numbers.
```

```python
le = LabelEncoder()
categorical_cols = ['department', 'region', 'education', 'gender', 'recruitment_channel']
```

```python
for col in categorical_cols:
    train_data[col] = le.fit_transform(train_data[col])
```

```python
#Scaling ensures all numeric features are on the same scale this is important for models like Logistic Regression, KNN, SVM.
from sklearn.preprocessing import StandardScaler
```

```python
x = train_data.drop('is_promoted', axis=1)
```

```python
y = train_data['is_promoted']
```

```python
scaler = StandardScaler()
x_scaled = pd.DataFrame(scaler.fit_transform(x), columns=x.columns)
```

```python
from sklearn.model_selection import train_test_split  #Train-Test Split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

# LINEAR REGRESSION MODEL

Linear Regression is used in HR analytics to predict numerical outcomes like salary or tenure, offering clear insights into how employee attributes affect these outcomes, which supports data-driven HR decisions.

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# RANDOM FOREST CLASSIFIER

Random Forest is used in HR analytics to build accurate, reliable models for classification tasks (like predicting attrition, promotion, or performance) by effectively handling complex, mixed-type, and noisy HR data.

```python
#random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)   #predict
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# Support Vector Machine

SVM is used in HR analytics to classify employees (like predicting attrition) accurately by effectively handling complex, multi-feature data and improving prediction reliability.

```python
from sklearn.svm import SVC
```

```python
#create an SVM classifier with RBF kernel (default)
svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')  # You can try 'linear' or 'poly' kernel too
svm_model.fit(x_train, y_train)
```

```python
svm_preds = svm_model.predict(x_test)
```

```python
print("SVM ccuracy:", accuracy_score(y_test, svm_preds))
```

```python
print("\nClassification Report:\n", classification_report(y_test, svm_preds))
```
